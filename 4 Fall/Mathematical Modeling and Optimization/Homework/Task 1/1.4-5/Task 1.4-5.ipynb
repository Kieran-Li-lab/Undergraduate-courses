# Nelder–Mead & DIRECT — Strict, Fast (Branin fix)
- **What changed vs. strict minimal**: DIRECT 仍按讲义伪代码实现，但增加两点**等价的加速**：
  1) **Memoized fu**：重复中心不会重复计算函数值（不改变算法轨迹）。
  2) **Coalesce equal d**：在潜在最优选择中，**相同半对角线长度 d** 只保留 **f 最小** 的一个点（其余在 (d,f) 平面上被支配，数学上可删）。
- 其它：域、ε=1e−6、kmax=150 均保持一致；输出与绘图保持不变。


```python
# Cell 1 — Imports & helpers
import numpy as np
import math, time, os
import matplotlib.pyplot as plt

plt.rcParams.update({"figure.dpi": 120})

def ensure_dir(path):
    d = os.path.dirname(path)
    if d and not os.path.exists(d):
        os.makedirs(d)

def savefig(path):
    ensure_dir(path)
    plt.savefig(path, bbox_inches="tight")
    plt.close()

def fmt_vec(v): return f"({v[0]:.6f}, {v[1]:.6f})"
def fmt_s(x):  return f"{x:.6f}"

def print_table(rows, header):
    print(header)
    for r in rows:
        print(f"{r[0]:10s} | {r[1]} | {r[2]:14s} | {fmt_vec(r[3])} | {fmt_s(r[4])} | {r[5]:10d} | {r[6]:7d} | {fmt_s(r[7])}")
```

```python
# Cell 2 — Test functions, domains, starts
def ackley(xy):
    x, y = xy
    term1 = -20.0 * math.exp(-0.2 * math.sqrt(0.5*(x*x + y*y)))
    term2 = -math.exp(0.5*(math.cos(2*math.pi*x)+math.cos(2*math.pi*y)))
    return term1 + term2 + math.e + 20.0

def branin(xy):
    x, y = xy
    a = 1.0; b = 5.1/(4*math.pi**2); c = 5.0/math.pi; r = 6.0; s = 10.0; t = 1.0/(8*math.pi)
    return a*(y - b*x*x + c*x - r)**2 + s*(1 - t)*math.cos(x) + s

def rosenbrock(xy):
    x, y = xy
    return (1.0 - x)**2 + 5.0*(y - x*x)**2

def rastrigin(xy):
    x, y = xy
    return 20.0 + (x*x - 10.0*math.cos(2*math.pi*x)) + (y*y - 10.0*math.cos(2*math.pi*y))

FUNS = {"ackley": ackley, "branin": branin, "rosenbrock": rosenbrock, "rastrigin": rastrigin}

# DIRECT domains — exact
BOUNDS_DIR = {
    "ackley":     (np.array([-2.0, -3.0]), np.array([4.0, 3.0])),
    "branin":     (np.array([ 0.0,  0.0]), np.array([4.0, 4.0])),
    "rosenbrock": (np.array([-5.0,  0.0]), np.array([2.0, 4.0])),
    "rastrigin":  (np.array([-1.0, -1.0]), np.array([6.0, 6.0]))
}

STARTS_NM = {
    "ackley":     np.array([0.0, 1.0]),
    "branin":     np.array([2.0, 2.0]),
    "rosenbrock": np.array([-1.5, 2.0]),
    "rastrigin":  np.array([2.5, 2.5])
}

KNOWN_MINIMA = {
    "ackley": [(0.0, 0.0)],
    "branin": [(math.pi, 2.275)],
    "rosenbrock": [(1.0, 1.0)],
    "rastrigin": [(0.0, 0.0)]
}
```

```python
# Cell 3 — Eval wrapper with counting
class FWrap:
    def __init__(self, f, name):
        self.f = f; self.name = name; self.evals = 0
    def __call__(self, x):
        self.evals += 1
        return float(self.f(np.array(x, float)))
```

```python
# Cell 4 — Nelder–Mead (unchanged from strict)
def nelder_mead(fw, x0, eps=1e-6, kmax=100, alpha=1.0, beta=2.0, gamma=0.5, delta=0.05):
    x0 = np.array(x0, float); n = len(x0)
    S = [x0.copy()]
    for i in range(n):
        e = np.zeros(n); e[i] = 1.0
        S.append(x0 + delta * e)
    S = np.array(S); y = np.array([fw(S[i]) for i in range(n+1)])
    it = 0; centroids = []; simplices = []
    while it < kmax and np.std(y) > eps:
        p = np.argsort(y); S = S[p]; y = y[p]
        xl, yl = S[0], y[0]; xh, yh = S[-1], y[-1]; xs, ys = S[-2], y[-2]
        xm = np.mean(S[:-1], axis=0); centroids.append(xm.copy()); simplices.append(S.copy())
        xr = xm + alpha*(xm - xh); yr = fw(xr)
        if yr < yl:
            xe = xm + beta*(xr - xm); ye = fw(xe)
            S[-1], y[-1] = (xe, ye) if ye < yr else (xr, yr)
        elif yr >= ys:
            if yr < yh:
                xh, yh = xr, yr; S[-1], y[-1] = xr, yr
            xc = xm + gamma*(xh - xm); yc = fw(xc)
            if yc > yh:
                for i in range(1, n+1):
                    S[i] = (S[i] + xl)/2.0; y[i] = fw(S[i])
            else:
                S[-1], y[-1] = xc, yc
        else:
            S[-1], y[-1] = xr, yr
        it += 1
    p = np.argsort(y); S = S[p]; y = y[p]
    xm = np.mean(S[:-1], axis=0); centroids.append(xm.copy()); simplices.append(S.copy())
    return {'x': S[0], 'f': y[0], 'iters': it, 'fevals': fw.evals,
            'centroids': np.array(centroids), 'simplices': simplices}
```

```python
# Cell 5 — DIRECT (strict math + faster engineering)
def _ab_from_unit(xu, a, b):  return a + xu * (b - a)
def _half_diag(depths):        return 0.5 * np.linalg.norm(3.0 ** (-np.array(depths, float)))

def direct(fw, a, b, eps=1e-6, kmax=150):
    a = np.array(a, float); b = np.array(b, float); n = len(a)
    assert n == 2, "2D only (as in assignment)"
    # memoization for fu at centers
    fu_cache = {}

    def fu(xu):
        # xu is small 2D array; hash by rounded tuple to avoid float-key noise
        key = (round(float(xu[0]), 15), round(float(xu[1]), 15))
        if key in fu_cache: return fu_cache[key]
        x = _ab_from_unit(xu, a, b)
        y = fw(x)
        fu_cache[key] = y
        return y

    # --- init ---
    c0 = np.full(n, 0.5)
    intervals = [[c0, fu(c0), [0]*n]]
    samples = [c0.copy()]
    fbest = intervals[0][1]; cbest = c0.copy()

    for _ in range(kmax):
        # build (d,f,idx) and **coalesce equal d** by keeping min f
        triples = []
        for idx, (c, y, depths) in enumerate(intervals):
            di = _half_diag(depths)
            triples.append((di, float(y), idx))
        # sort by d asc, f asc
        triples.sort(key=lambda t: (t[0], t[1]))
        # coalesce: for equal d within tol, keep minimal f only
        coalesced = []
        tol = 1e-14
        i = 0
        while i < len(triples):
            d_i = triples[i][0]
            j = i
            f_min, idx_min = triples[i][1], triples[i][2]
            while j + 1 < len(triples) and abs(triples[j+1][0] - d_i) < tol:
                j += 1
                if triples[j][1] < f_min:
                    f_min, idx_min = triples[j][1], triples[j][2]
            coalesced.append((d_i, f_min, idx_min))
            i = j + 1

        # lower convex hull on coalesced set
        hull = []
        for (d, f, idx) in coalesced:
            while len(hull) >= 2:
                d1, f1, i1 = hull[-1]
                d2, f2, i2 = hull[-2]
                if abs(d1 - d2) < tol:
                    if f1 >= f: hull.pop()
                    else: break
                    continue
                if abs(d - d1) < tol:
                    if f1 <= f: break
                    else: hull.pop(); continue
                slope1 = (f1 - f2) / (d1 - d2)
                slope2 = (f - f1) / (d - d1)
                if slope2 >= slope1 - 1e-18:
                    hull.pop()
                else:
                    break
            hull.append((d, f, idx))

        thr = fbest + eps * abs(fbest)
        sel = [intervals[idx] for (d, f, idx) in hull if f <= thr]
        if not sel:
            best_idx = min(range(len(intervals)), key=lambda j: intervals[j][1])
            sel = [intervals[best_idx]]

        # remove selected (id-based) and subdivide along min-depth dims
        S_ids = set(id(iv) for iv in sel)
        keep = [iv for iv in intervals if id(iv) not in S_ids]

        new_list = []
        for (c, y, depths) in sel:
            depths = list(depths)
            dmin = min(depths)
            I = [i for i, di in enumerate(depths) if di == dmin]
            # steps and children
            for i_dim in I:
                step = 3.0 ** (-(depths[i_dim] + 1))
                e = np.zeros_like(c); e[i_dim] = 1.0
                c1 = c + step * e; c2 = c - step * e
                y1 = fu(c1); y2 = fu(c2)
                d_new = depths.copy(); d_new[i_dim] += 1
                new_list.append([c1, y1, d_new])
                new_list.append([c2, y2, d_new])
                samples.append(c1.copy()); samples.append(c2.copy())
                if y1 < fbest: fbest, cbest = y1, c1.copy()
                if y2 < fbest: fbest, cbest = y2, c2.copy()
            # keep the center with updated depths (+1 on all split dims)
            d_keep = depths.copy()
            for i_dim in I: d_keep[i_dim] += 1
            new_list.append([c, y, d_keep])

        intervals = keep + new_list

    x_best = _ab_from_unit(cbest, a, b)
    return {'x': x_best, 'f': float(fbest), 'iters': kmax, 'fevals': fw.evals, 'samples': np.array(samples)}
```

```python
# Cell 6 — Plotting (equal-aspect, square window)
def _square_window(xy_list, pad_ratio=0.10):
    xs = [p[0] for p in xy_list]; ys = [p[1] for p in xy_list]
    x_min, x_max = min(xs), max(xs); y_min, y_max = min(ys), max(ys)
    span = max(max(x_max - x_min, 1e-9), max(y_max - y_min, 1e-9))
    cx, cy = 0.5*(x_min + x_max), 0.5*(y_min + y_max)
    pad = pad_ratio * span
    return cx - 0.5*span - pad, cx + 0.5*span + pad, cy - 0.5*span - pad, cy + 0.5*span + pad

def _contour(fun, xmin, xmax, ymin, ymax, N=120):
    X = np.linspace(xmin, xmax, N); Y = np.linspace(ymin, ymax, N)
    XX, YY = np.meshgrid(X, Y); ZZ = np.zeros_like(XX)
    for i in range(N):
        for j in range(N):
            ZZ[i,j] = fun(np.array([XX[i,j], YY[i,j]]))
    plt.contour(XX, YY, ZZ, levels=30)

def plot_nm(fun, res, fname, minima):
    pts = []
    for S in res['simplices']:
        pts.extend(S.tolist())
    pts.append(res['x'].tolist())
    for (mx,my) in minima: pts.append([mx,my])
    xmin, xmax, ymin, ymax = _square_window(pts, pad_ratio=0.12)
    plt.figure(); _contour(fun, xmin, xmax, ymin, ymax, N=120)
    for S in res['simplices']:
        P = np.vstack([S, S[0]]); plt.plot(P[:,0], P[:,1], marker='o', linewidth=1, alpha=0.8)
    C = res['centroids']; plt.plot(C[:,0], C[:,1], marker='x', linewidth=1.5, label='centroid')
    plt.scatter([res['x'][0]], [res['x'][1]], marker='*', s=80, label='final')
    for (mx,my) in minima: plt.scatter([mx], [my], marker='+', s=60, label='global min')
    ax = plt.gca(); ax.set_aspect('equal', adjustable='box'); ax.set_xlim(xmin, xmax); ax.set_ylim(ymin, ymax)
    plt.legend(); plt.title('Nelder–Mead'); plt.xlabel('x'); plt.ylabel('y'); plt.savefig(fname, bbox_inches='tight'); plt.close()

def plot_direct(fun, res, fname, a, b, minima):
    su = res['samples']; so = a + su*(b - a)
    pts = so.tolist() + [res['x'].tolist()] + [list(m) for m in minima]
    xmin, xmax, ymin, ymax = _square_window(pts, pad_ratio=0.12)
    plt.figure(); _contour(fun, xmin, xmax, ymin, ymax, N=120)
    plt.scatter(so[:,0], so[:,1], s=10, alpha=0.6, label='sampled centers')
    plt.scatter([res['x'][0]], [res['x'][1]], marker='*', s=80, label='best found')
    for (mx,my) in minima: plt.scatter([mx], [my], marker='+', s=60, label='global min')
    ax = plt.gca(); ax.set_aspect('equal', adjustable='box'); ax.set_xlim(xmin, xmax); ax.set_ylim(ymin, ymax)
    plt.legend(); plt.title('DIRECT samples'); plt.xlabel('x'); plt.ylabel('y'); plt.savefig(fname, bbox_inches='tight'); plt.close()
```

```python
# Cell 7 — Run experiments (save plots + print results)
rows = []
eps = 1e-6; kmax_nm = 100; kmax_dir = 150

for name in ['ackley','branin','rosenbrock','rastrigin']:
    f = FUNS[name]

    # Nelder–Mead
    x0 = STARTS_NM[name]
    fw_nm = FWrap(f, name + '_nm')
    t0 = time.time(); res_nm = nelder_mead(fw_nm, x0, eps=eps, kmax=kmax_nm, delta=0.05); t1 = time.time()
    plot_nm(f, res_nm, f"plot_{name}_nelder_trajectory.png", KNOWN_MINIMA[name])
    rows.append((name, (float(x0[0]), float(x0[1])), 'Nelder–Mead', res_nm['x'], res_nm['f'], res_nm['iters'], fw_nm.evals, t1-t0))

    # DIRECT
    a, b = BOUNDS_DIR[name]
    fw_dir = FWrap(f, name + '_dir')
    t0 = time.time(); res_dir = direct(fw_dir, a, b, eps=eps, kmax=kmax_dir); t1 = time.time()
    plot_direct(f, res_dir, f"plot_{name}_direct_samples.png", a, b, KNOWN_MINIMA[name])
    rows.append((name, (tuple(a.tolist()), tuple(b.tolist())), 'DIRECT', res_dir['x'], res_dir['f'], res_dir['iters'], fw_dir.evals, t1-t0))

print_table(rows, "function   | start           | method         | x found                 | f(x found) | iterations | f evals | time s")
```

```python
#Cell 8 — Comparative check (final f, evals, time)
by_func = {}
for r in rows:
    by_func.setdefault(r[0], []).append(r)

print("function   | f(NM)       | evals(NM) | time(NM) | f(DIRECT)  | evals(DIR) | time(DIR)")
for name in ['ackley','branin','rosenbrock','rastrigin']:
    nm = next(rr for rr in by_func[name] if rr[2].startswith('Nelder'))
    dr = next(rr for rr in by_func[name] if rr[2].startswith('DIRECT'))
    print(f"{name:10s} | {nm[4]:.6e} | {nm[6]:9d} | {nm[7]:.4f} | {dr[4]:.6e} | {dr[6]:9d} | {dr[7]:.4f}")
```
